# numpy.cov

Estimate a covariance matrix, given data and weights.

### Default
```
numpy.cov(m, y=None, rowvar=True, bias=False, ddof=None, fweights=None, aweights=None)
```

### Parameters
#### bias
Default normalization (False) is by (N - 1), where N is the number of observations given (unbiased estimate). If bias is True, then normalization is by N.

````python
import numpy as np
````

````python
x = np.array([[0, 2], [1, 1], [2, 0]]).T
x
````

````python
np.cov(x)
````

````python
x = [-2.1, -1,  4.3]
y = [3,  1.1,  0.12]
X = np.stack((x, y), axis=0)
X
````

````python
X.shape
````

````python
np.cov(X)
````

````python
np.cov(x, y)
````

## np.covを用いてランダムな値を生成

````python
import matplotlib.pyplot as plt
````

````python
x = [-2.1, -1,  4.3]
y = [3,  1.1,  0.12]
X = np.stack((x, y), axis=0)
cov = np.cov(X)
cov
````

````python
mean = np.array([0, 10])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x');
````

### covarianceが１の時

````python
cov = np.array([[1,1], [1,1]])
mean = np.array([0, 10])
x, y = np.random.multivariate_normal(mean, cov, 5000).T
plt.plot(x, y, 'x');
````
